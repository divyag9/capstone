```python
# using regex to remove any character that is not alpha numeric or punctuation
pattern = r"[^{}]".format('a-zA-Z0-9!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~') 
clean_text = re.sub(pattern,' ',test_txt)
```

```python
doc = nlp(clean_text)
```

```python
for token in doc:
    print(token.text, token.pos_)
```

```python
# retain only the nouns, pronouns and adjectives
# get the vector representations for the words - right now using spacy which uses glove
tokens = []
embedings = []
pos_to_retain = {'NOUN','ADJ','PROPN'}
for token in doc:
    if token.pos_ in pos_to_retain:
        tokens.append(token.text)
        embedings.append(token.vector)
        
sentence_embedding = doc.vector 
```

```python
# try to use bert, by getting the embeddings first for sentence and then retaining only nouns,adj,propn
```

```python
# compute the similarity scores to put in a matrix using cosine similarity
num_tokens = len(tokens)
simlarity_matrix = np.zeros([num_tokens, num_tokens])
for i in range(num_tokens):
    for j in range(num_tokens):
        if i != j:
            simlarity_matrix[i][j] = cosine_similarity(embedings[i].reshape(1,300), embedings[j].reshape(1,300))[0,0]
```

```python
# compute page rank scores
nx_graph = nx.from_numpy_array(simlarity_matrix)
scores = nx.pagerank(nx_graph)
```

```python
CONVERGENCE_THRESHOLD = 0.001
def pagerank_weighted(graph, initial_value=None, damping=0.85):
    """Calculates PageRank for an undirected graph"""
    if initial_value == None: initial_value = 1.0 / len(graph.nodes())
    scores = dict.fromkeys(graph.nodes(), initial_value)

    iteration_quantity = 0
    for iteration_number in range(100):
        iteration_quantity += 1
        convergence_achieved = 0
        print('iter_num:', iteration_number)
        for i in graph.nodes():
            rank = 1 - damping
            for j in graph.neighbors(i):
                neighbors_sum = sum(graph.edges[j, k]['weight'] for k in graph.neighbors(j))
                rank += damping * scores[j] * graph.edges[j, i]['weight'] / neighbors_sum

            if abs(scores[i] - rank) <= CONVERGENCE_THRESHOLD:
                convergence_achieved += 1

            scores[i] = rank
            print('i, rank:', i, ',',rank)
        if convergence_achieved == len(graph.nodes()):
            break

    return scores
```

```python
import pke

# initialize keyphrase extraction model, here TopicRank
extractor = pke.unsupervised.TopicRank()

# load the content of the document, here document is expected to be in raw
# format (i.e. a simple text file) and preprocessing is carried out using spacy
extractor.load_document(input='Text.txt', language='en')

# keyphrase candidate selection, in the case of TopicRank: sequences of nouns
# and adjectives (i.e. `(Noun|Adj)*`)
extractor.candidate_selection()

# candidate weighting, in the case of TopicRank: using a random walk algorithm
extractor.candidate_weighting()

# N-best selection, keyphrases contains the 10 highest scored candidates as
# (keyphrase, score) tuples
keyphrases = extractor.get_n_best(n=20)
```
