```python
from textblob import TextBlob 

blob1 = TextBlob(e1)

print(blob1.tags)           
```

```python
#sentiment analysis

testimonial = TextBlob(e1)

print(testimonial.sentiment)

print(testimonial.sentiment.polarity)
```

```python
# Parser
print(blob1.parse())
```

```python
#Diff method of detecting sentiment

from textblob.sentiments import NaiveBayesAnalyzer

blobS = TextBlob(e1, analyzer=NaiveBayesAnalyzer())
blobS.sentiment
```

```python
# Noun Phrase Chunking

from textblob.np_extractors import ConllExtractor
import nltk
nltk.download('conll2000')

extractor = ConllExtractor()
blob = TextBlob(e1, np_extractor=extractor)
blob.noun_phrases
```

```python
#can specify NLTK POS tagging if desired

from textblob.taggers import NLTKTagger

nltk_tagger = NLTKTagger()
blobPOS = TextBlob(e1, pos_tagger=nltk_tagger)
blobPOS.pos_tags
```
