```python
def cleanText(text):
    
    def replace_contraction(text):
        contraction_patterns = [(r'won\'t', 'will not'), (r'can\'t', 'can not'), (r'i\'m', 'i am'), (r'ain\'t', 'is not'), (r'(\w+)\'ll', '\g<1> will'), (r'(\w+)n\'t', '\g<1> not'),
                                (r'(\w+)\'ve', '\g<1> have'), (r'(\w+)\'s', '\g<1> is'), (r'(\w+)\'re', '\g<1> are'), (r'(\w+)\'d', '\g<1> would'), (r'&', 'and'), (r'dammit', 'damn it'), (r'dont', 'do not'), (r'wont', 'will not') ]
        patterns = [(re.compile(regex), repl) for (regex, repl) in contraction_patterns]
        for (pattern, repl) in patterns:
            (text, count) = re.subn(pattern, repl, text)
        return text

    def replace_links(text, filler=' '):
        text = re.sub(r'((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*',
                          filler, text).strip()
        return text

    def remove_numbers(text):
        text = ''.join([i for i in text if not i.isdigit()])
        return text
    
    text = text.strip().replace("\n", " ").replace("\r", " ")
    text = replace_contraction(text)
    text = replace_links(text, "link")
    #text = remove_numbers(text)
    text = re.sub(r'[,!@#$%^&*)(|/><";:.?\'\\}{]',"",text)
    text = text.lower()
    return text
```

```python
df['clean_captions'] = df['captions'].apply(cleanText)
```

```python
def sentKeywords(text):
    key_words = []
    amod_dict = {}
    skip_set = set()
    doc = nlp(text)
    for token in doc:

        if token.i in skip_set:
            continue
            
        if (token.dep_ == 'amod' and token.pos_ == 'ADJ' and token.head.dep_ == 'nsubj'):
            amod_dict[token.head.i] = token.i

        if token.dep_ == 'compound' and (token.pos_ == 'NOUN' or token.pos_ == 'PROPN'):
            kw_list = []
            x = token
            while x.dep_ == 'compound' and (x.pos_ == 'NOUN' or x.pos_ == 'PROPN'):
                skip_set.add(x.i)
                kw_list.append(x.i)
                if x.head.dep_ == 'nsubj':
                    if x.head.i in amod_dict:
                        kw_list.append(amod_dict[x.head.i])
                    kw_list.append(x.head.i)
                    skip_set.add(x.head.i)
                x = x.head
                
            kw_list.sort()
            kw = ''
            cnt = 0
            for k in kw_list:
                if cnt == 0:
                    cnt += 1
                    kw += doc[k].text
                else:
                    kw += ' '
                    kw += doc[k].text
            key_words.append(kw)

        if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
            kw_list = []
            if token.i in amod_dict:
                kw_list.append(amod_dict[token.i])
            kw_list.append(token.i)
            kw_list.sort()
            kw = ''
            cnt = 0
            for k in kw_list:
                if cnt == 0:
                    cnt += 1
                    kw += doc[k].text
                else:
                    kw += ' '
                    kw += doc[k].text
            key_words.append(kw)
                
    return key_words
```
